### Usando o MatplotLib

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
# Usando o format string para mudar a visualização do eixo Y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
# Adicionando um titulo
plt.title("Valor total gasto por unidade da federação em agosto de 2008")
# mostra somente o gráfico
plt.show()
```

#### Desafio 02:

```python
dados["2021/Ago"] = dados["2021/Ago"].replace("-", "0", regex=True, )
dados["2021/Ago"] = pd.to_numeric(dados["2021/Ago"], downcast="float")
dados["2021/Ago"]
```

```python
dados.sort_values(by=['2021/Ago'], ascending = True, inplace=True)
ax = dados.plot.barh(x="Unidade da Federação", y='2021/Ago',figsize=(15,8), color='darkgreen')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
# ax.set_facecolor('tab:gray')
ax.patch.set_facecolor('black')
ax.set_ylabel('')
ax.xaxis.grid()
plt.title("Valor Gasto SUS por Unidade da Federação em Agosto de 2021")
ax.get_legend().remove()
plt.show()
```

## 2.Primeiras visualizações de dados

### Entendendo os gráficos

```python
# Fazendo um slice dos dados (fatiando)
dados[['2008/Ago','2008/Set']].head()
```

```python
# Selecionando somente os meses que tem todos os valores
# Usando o mean buscamos somente os meses com valores, ignorando os que estão com algum valor faltando
dados_validos = dados.mean().index
dados_validos = dados_validos.insert(0, "Unidade da Federação")
dados_validos
```

```python
usaveis = dados[dados_validos]
usaveis.head()
```

```python
# Setando o nome dos estados como index
usaveis = usaveis.set_index("Unidade da Federação")
usaveis.head()
```

```python
# T -> Inverte a Coluna e Linhas
usaveis.T.head()
```

```python
usaveis.T.plot(figsize=(10,7))
# Gerou o gráfico com todos os estados porém no final há um valor muito grande
```

```python
# Removendo a coluna Total
usaveis = usaveis.drop('Total', axis=1)
```

```python
usaveis.T.plot(figsize=(10,7))
```

### Desafios

#### Desafio 01:

Reposicionar a legenda do gráfico em uma posição mais adequada

```python
ax = usaveis.T.plot(figsize=(10,7))
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos por ano/mes segundo Unidade da Federação")
plt.show()
```
