```python
dados["2021/Ago"] = dados["2021/Ago"].replace("-", "0", regex=True, )
dados["2021/Ago"] = pd.to_numeric(dados["2021/Ago"], downcast="float")
dados["2021/Ago"]
```

```python
dados.sort_values(by=['2021/Ago'], ascending = True, inplace=True)
ax = dados.plot.barh(x="Unidade da Federação", y='2021/Ago',figsize=(15,8), color='darkgreen')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
# ax.set_facecolor('tab:gray')
ax.patch.set_facecolor('black')
ax.set_ylabel('')
ax.xaxis.grid()
plt.title("Valor Gasto SUS por Unidade da Federação em Agosto de 2021")
ax.get_legend().remove()
plt.show()
```

```python
# Removendo a coluna Total
usaveis = usaveis.drop('Total', axis=1)
```
